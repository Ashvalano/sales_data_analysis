```python
import glob
import os
import re

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Merge 12 months of sales data into one

```python

csv_files = glob.glob('Sales_Data/Sales_*.csv')
# print(csv_files)

dataframes = []

for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)
    
merged_df = pd.concat(dataframes, ignore_index=True)
merged_df.to_csv('merged_data.csv', index=False)

merged_df.head(10)
```

### Data Cleaning

```python
# merged_df.loc[merged_df['Order Date'].isnull()]
# nan_df = merged_df[merged_df.isnull().any(axis=1)]
# nan_df.head()

# drop nan values
merged_df = merged_df.dropna(how='all')

# Remove rows with placeholder values
merged_df = merged_df[merged_df['Order Date'] != 'Order Date'] 

# see how many null values there are
print(merged_df.isnull().sum())
merged_df.head()
```

### Add month and sales column, visualize by monthly sales

```python

# reset index after setting
# merged_df.reset_index(inplace=True)

# find specific element using loc and iloc
# merged_df.loc[(merged_df['Order ID']== '177057')]
# print(merged_df.iloc[519]) # index 519

# converted to datetime obj
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%m/%d/%y %H:%M')

# get month value
merged_df['Month'] = pd.to_datetime(merged_df['Order Date']).dt.month
merged_df['Month'] = merged_df['Order Date'].dt.strftime('%B')
# merged_df.drop('Month', axis =1)

merged_df['Quantity Ordered'] = pd.to_numeric(merged_df['Quantity Ordered'])
merged_df['Price Each'] = pd.to_numeric(merged_df['Price Each'])
merged_df['Sales'] = merged_df['Quantity Ordered'] * merged_df['Price Each']

# arrange month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
              'October','November','December']
merged_df['Month'] = pd.Categorical(merged_df['Month'], categories=month_order, ordered=True)


monthly_sales = merged_df.groupby('Month')['Sales'].sum()

fig, ax = plt.subplots(figsize=(8, 6))

# Plot the monthly sales data
ax.bar(monthly_sales.index, monthly_sales.values)
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Monthly Sales')

# Customize the appearance of the plot
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()

# for double checking data
print(monthly_sales)
```

### Add city column, sales by city, visualize

```python

# str.strip() is not necessary, but is for removing leading or trailing whitespace

# drop multiple columns
# merged_df.drop(['City','State'], axis=1, inplace=True)

# get city value from purchase address
merged_df['City'] = merged_df['Purchase Address'].str.split(',').str[1:3].apply(
    lambda x: ', '.join([s.strip() for s in x]))

# remove postal code
merged_df['City'] = merged_df['City'].str.replace(r'\b\d+\b', '', regex=True).str.strip()


# visualizing
sales_by_city = merged_df.groupby('City')['Sales'].sum()

fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data
ax.bar(sales_by_city.index, sales_by_city.values)
ax.set_xlabel('City')
ax.set_ylabel('Sales')
ax.set_title('Sales by City')

# Customize the appearance of the plot, rotate city names 45 degrees
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()

# for double checking data
print(sales_by_city)
```

### Most sales count by time of day

```python

# get hour value
merged_df['Hour'] = merged_df['Order Date'].dt.hour
hours = merged_df['Hour'].value_counts().sort_index().index.tolist()

# plot
plt.plot(hours, merged_df.groupby(['Hour']).count(), marker='o', linestyle='-', color='blue')
plt.xticks(hours)
plt.xlabel('Hour of Day')
plt.ylabel('Sales Count')

plt.grid()
plt.figure(figsize=(12, 6))
plt.show
```

### Sales count by product

```python
# sales count by product
product_sold_count = merged_df.groupby('Product')['Quantity Ordered'].sum()

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.bar(product_sold_count.index, product_sold_count.values, label='Number of Products Sold')

ax.set_xlabel('Product')
ax.set_ylabel('Sales Count')
ax.set_title('Number of Products Sold')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()
```

### Product mostly sold and their prices

```python

# prod = merged_df['Product'].unique()
# print(prod)

# products sold count and their prices
product_sold_count = merged_df.groupby('Product')['Quantity Ordered'].sum()
product_price = merged_df.groupby('Product')['Price Each'].mean()

fig, ax = plt.subplots(figsize=(12, 6))
ax2 = ax.twinx()

# Plot sales count of products as bar
ax.bar(product_sold_count.index, product_sold_count.values, label='Number of Products Sold')

# Plot product prices as line
ax2.plot(product_price.index, product_price.values, marker='o', color='red', label='Price')

ax.set_xlabel('Product')
ax.set_ylabel('Sales Count', color='blue')
ax2.set_ylabel('Price', color='red')
ax.set_title('Number of Products Sold and their Prices')

ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# cant fix xticktabels
# ax.set_xticks(range(len(product_sold_count.index)))
# ax.set_xticklabels(product_sold_count.index, rotation=45, fontsize=8, ha='right')
ax2.set_xticks([]) 

plt.xticks(rotation=45,fontsize=8, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()
```

### Products most often sold together

```python
df = merged_df[merged_df.duplicated('Order ID', keep=False)]

df['Grouped Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
df = df[['Order ID', 'Grouped Products']].drop_duplicates()

grouped_counts = df['Grouped Products'].value_counts()
top_products = grouped_counts.head(10)

plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')

plt.xlabel('Grouped Products')
plt.ylabel('Count')
plt.title('Top Grouped Products by Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')
plt.show()

df.head()
print(top_products)
```
